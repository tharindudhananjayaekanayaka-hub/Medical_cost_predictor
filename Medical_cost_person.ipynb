```python
# define a x and y . y is charges and x is another features
x = df.drop(columns='charges', axis=1)
y = df['charges']
```

```python
# print in a categorical features and numerical features
categorical_features = x.select_dtypes(include='object').columns
numerical_features = x.select_dtypes(exclude='object').columns
print(categorical_features)
print(numerical_features)
```

```python
# dataset devide a split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# train shape and test shape
print(x_train.shape)
print(x_test.shape)
```

```python
# preprocessing part categorical features using OneHotEncoder, numerical features using a Standerscale
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features),
        ('num', StandardScaler(), numerical_features)
    ])

```

```python
# using a linear regression pipline using
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression()) # Changed LogisticRegression to LinearRegression
])
# model fit
pipeline.fit(x_train, y_train)
```

```python
# linear regression using mtrics
y_pred = pipeline.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
RMSE = np.sqrt(mse)
print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {RMSE}')
print(f'R-squared: {r2}')
```

Desitiontreeclassifier

```python
# decision tree pipline using
from sklearn.tree import DecisionTreeRegressor # Changed from DecisionTreeClassifier
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', DecisionTreeRegressor()) # Changed to regressor and DecisionTreeRegressor
])
# model fit
pipeline.fit(x_train, y_train)
```

```python
# model using a mtrics
```

```python
# Decision Tree Regressor using metrics
y_pred_dt = pipeline.predict(x_test)
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)
rmse_dt = np.sqrt(mse_dt)
print(f'Decision Tree Mean Squared Error: {mse_dt}')
print(f'Decision Tree Root Mean Squared Error: {rmse_dt}')
print(f'Decision Tree R-squared: {r2_dt}')
```

Randomforest


```python
# using a randomforestRegressor pipline
from sklearn.ensemble import RandomForestRegressor
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor())
])
# model fit
pipeline.fit(x_train, y_train)
```

```python
# using a Matrics in randomforeregression
y_pred_rf = pipeline.predict(x_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
print(f'Random Forest Mean Squared Error: {mse_rf}')
print(f'Random Forest Root Mean Squared Error: {rmse_rf}')
print(f'Random Forest R-squared: {r2_rf}')
```

```python
# all include model performance in matrics
models = pd.DataFrame({
    'Model': ['Linear Regression', 'Decision Tree', 'Random Forest'],
    'R2 Score': [r2, r2_dt, r2_rf],
    'MSE': [mse, mse_dt, mse_rf],
    'RMSE': [RMSE, rmse_dt, rmse_rf]
})
print("Model Performance:")
print(models)
```

```python
# in this best_model_pipline save in a joblib file
best_model_pipeline = pipeline
joblib.dump(best_model_pipeline, 'best_model_pipeline.joblib')
```
